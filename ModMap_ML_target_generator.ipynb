```python
for x in range(len(data)):
    for y in range(len(data)):
        if data.iloc[x,y]>=((miu-sigma)*1):
            continue
        else:
            ll = x-4
            if ll<0:
                ll=0
            elif ll>256:
                ll=256
            rl = x+4
            if rl<0:
                rl=0
            elif rl>256:
                rl=256
            ul = y+4
            if ul<0:
                ul=0
            elif ul>256:
                ul=256
            bl = y-4
            if bl<0:
                bl=0
            elif bl>256:
                bl=256
            grid = data.iloc[ll:rl,bl:ul]
            Sum=0
            count=0
            for i in range(len(grid)):
                for j in range(len(grid.T)):
                    if grid.iloc[i,j] >= (miu-sigma):
                        Sum += grid.iloc[i,j]
                        count += 1
                    else: 
                        Sum +=0
                        count +=0
                if Sum==0:
                    data.iloc[x,y] = data.values.mean()
                else:
                    print(Sum/count)
                    data.iloc[x,y] = Sum/count
        
```
